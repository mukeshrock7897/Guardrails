# ✅ 1.7 Guardrails in LLM Applications

## 🤖 Meaning

**Guardrails in LLM applications control how the AI reads input, uses context, generates answers, and returns safe output.**

---

## 🛡️ Used For

```text id="7obc0k"
Checking user input
Protecting private data
Controlling retrieved context
Reducing hallucinations
Validating final response
Following business rules
```

---

## 🏦 Finance Example

In a finance chatbot, guardrails make sure the LLM does not:

```text id="w81pqe"
Reveal client data
Give guaranteed returns
Use unapproved sources
Make unsupported claims
Ignore disclaimers
```
