# ✅ 1.1 What are AI Guardrails

## 🧠 Definition

**AI Guardrails are rules and checks that control an AI system so it gives safe, accurate, secure, and allowed responses.**

---

## 🎯 Why It Matters

Guardrails help AI avoid:

```text
Wrong answers
Unsafe responses
Data leakage
Policy violations
Unauthorized actions
Hallucinations
```

---

## 🧱 Simple Flow

```text
User Input
   ↓
Input Check
   ↓
AI Response
   ↓
Output Check
   ↓
Safe Final Answer
```

---

## 🏦 Finance Example

In finance, guardrails stop AI from:

```text
Giving guaranteed investment advice
Leaking client portfolio data
Using unapproved research
Making false market claims
Executing trades without approval
```

## ✅ One-Line Summary

**AI Guardrails are safety layers that check input, output, data access, and actions before AI gives the final response.**
