# ✅ 1.8 Guardrails in Agentic AI Systems

## 🧭 Meaning

**Guardrails in Agentic AI control what an AI agent can think, decide, and do using tools, APIs, files, or databases.**

---

## 🛡️ Used For

```text id="hf7mwl"
Tool access control
API action validation
Database safety
Human approval
Risky action blocking
Audit logging
```

---

## 🏦 Finance Example

In a finance AI agent, guardrails stop the agent from:

```text id="qfx1zh"
Executing trades without approval
Accessing restricted client data
Calling unauthorized APIs
Changing portfolio data directly
Sending risky investment advice
```
