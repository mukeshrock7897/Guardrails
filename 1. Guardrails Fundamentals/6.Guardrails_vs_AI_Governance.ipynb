# ✅ 1.6 Guardrails vs AI Governance

## ⚖️ Simple Difference

**AI Governance defines rules and policies.**
**Guardrails enforce those rules inside the AI system.**

---

## ⚖️ AI Governance Handles

```text
AI policies
Risk standards
Compliance rules
Approval process
Audit requirements
Responsible AI principles
```

---

## 🛡️ Guardrails Handle

```text
Input checks
Output checks
Data access control
Tool/action control
Policy enforcement
Human escalation
```

---

## 🏦 Finance Example

### AI Governance

```text
Company policy says AI must not give unauthorized investment advice.
```

### Guardrails

```text
System blocks or rewrites answers that look like direct investment advice.
```

---

## ⭐ Key Difference

| AI Governance                  | Guardrails                    |
| ------------------------------ | ----------------------------- |
| Defines rules                  | Enforces rules                |
| High-level process             | Technical implementation      |
| Owned by risk/compliance teams | Built by AI/engineering teams |

---

## ✅ One-Line Summary

**AI Governance decides what AI should follow; guardrails make sure the AI actually follows it.**
