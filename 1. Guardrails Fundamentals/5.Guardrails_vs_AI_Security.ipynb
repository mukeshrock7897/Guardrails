# ✅ 1.5 Guardrails vs AI Security

## 🔐 Simple Difference

**AI Security protects the AI system from attacks.**
**Guardrails control what the AI is allowed to do and say.**

---

## 🔐 AI Security Handles

```text
Authentication
Authorization
API security
Data protection
Network security
Model access security
```

---

## 🛡️ Guardrails Handle

```text
Prompt injection
Unsafe responses
Hallucinations
Policy violations
Tool misuse
Sensitive data leakage
```

---

## 🏦 Finance Example

### AI Security

```text
Only authorized employees can access the finance AI app
```

### Guardrails

```text
Even authorized users cannot ask AI to reveal restricted client portfolio data
```

---

## ⭐ Key Difference

| AI Security                  | Guardrails                         |
| ---------------------------- | ---------------------------------- |
| Protects system access       | Controls AI behavior               |
| Focuses on infrastructure    | Focuses on AI input/output/actions |
| Prevents unauthorized access | Prevents unsafe AI behavior        |

---

## ✅ One-Line Summary

**AI Security protects the system; guardrails control the AI’s behavior inside the system.**
