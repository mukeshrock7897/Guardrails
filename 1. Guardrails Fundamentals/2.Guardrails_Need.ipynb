# ✅ 1.2 Why Guardrails are Needed

## 🎯 Main Need

**Guardrails are needed to keep AI safe, controlled, accurate, and suitable for real-world production use.**

---

## 🚫 Without Guardrails, AI Can

```text
Leak sensitive data
Give wrong answers
Follow malicious prompts
Break company policies
Take risky actions
Generate non-compliant responses
```

---

## 🛡️ With Guardrails, AI Can

```text
Block unsafe inputs
Protect private data
Reduce hallucinations
Follow business rules
Control tool/API actions
Escalate risky cases to humans
```

---

## 🏦 Finance Example

Guardrails are needed so AI does not:

```text
Give guaranteed profit advice
Expose client portfolio details
Use unapproved financial data
Trigger trades without approval
Ignore compliance rules
```
