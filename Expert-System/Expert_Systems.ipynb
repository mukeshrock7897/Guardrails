# **Beginner Level**

**1. Introduction to Expert Systems**

* An expert system is a computer program that mimics the decision-making abilities of a human expert.

**History:**
* Originated in the 1970s, with notable systems like DENDRAL and MYCIN.

**Components:**
* **Knowledge Base:** Stores all the facts and rules.
* **Inference Engine:** Applies logical rules to the knowledge base to deduce new information.
* **User Interface:** Allows users to interact with the system.
* **Explanation Facility:** Explains the reasoning process to the user.
* **Knowledge Acquisition Module:** Helps in updating the knowledge base.

**Applications:**
* Medical diagnosis, financial services, customer support, etc.

**Advantages:**
* Consistency , availability, can handle large amounts of information.

**Disadvantages:**
* Lack of common sense, high development cost, difficulty in updating.

```python
# Simple rule-based system example
def diagnose_symptom(symptom):
    if symptom == "fever":
        return "You might have the flu."
    elif symptom == "headache":
        return "You might have a migraine."
    else:
        return "Symptom not recognized."

print(diagnose_symptom("fever"))
```

```python
# Example of forward chaining
rules = {
    "rain": "wet ground",
    "wet ground": "slippery road",
}

facts = ["rain"]
for fact in facts:
    if fact in rules:
        facts.append(rules[fact])

print(facts)
```

```python
import numpy as np

# Example of simple fuzzy logic
def fuzzy_temp(temp):
    if temp <= 20:
        return "Cold"
    elif 20 < temp <= 25:
        return "Warm"
    else:
        return "Hot"

print(fuzzy_temp(22))
```

**9.Case-Based Reasoning**

* **Case Libraries:** Storing and retrieving past cases to solve new problems.
* **Case Matching and Adaptation:** Finding the closest match and adapting the solution

```python
cases = {
    "case1": {"symptom": "fever", "diagnosis": "flu"},
    "case2": {"symptom": "headache", "diagnosis": "migraine"},
}

def diagnose_case(symptom):
    for case in cases.values():
        if case["symptom"] == symptom:
            return case["diagnosis"]
    return "No matching case found."

print(diagnose_case("fever"))
```
