# ✅ 2.3 Chatbot Safety Filters

## 🤖 Meaning

**Chatbot Safety Filters are checks used to stop chatbots from giving unsafe, harmful, or unwanted responses.**

---

## 🛡️ Used For

```text
Blocking unsafe replies
Avoiding harmful suggestions
Filtering abusive content
Preventing policy violations
Keeping chatbot behavior controlled
```

---

## 🏦 Finance Example

A finance chatbot safety filter can stop replies like:

```text
This stock will 100% give profit.
Share your client login password.
Ignore compliance and execute the trade.
```
