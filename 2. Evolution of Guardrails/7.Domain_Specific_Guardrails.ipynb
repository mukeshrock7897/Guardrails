# ✅ 2.7 Domain-Specific Guardrails

## 🏦 Meaning

**Domain-Specific Guardrails are rules created for a particular industry, business area, or use case.**

---

## 🛡️ Used For

```text
Following industry rules
Controlling domain-specific risks
Using approved data sources
Avoiding wrong expert advice
Protecting sensitive business data
```

---

## 🏦 Finance Example

In finance, guardrails control:

```text
Investment advice
Client portfolio data
Market research sources
Risk disclosures
Trade approvals
Compliance checks
```
