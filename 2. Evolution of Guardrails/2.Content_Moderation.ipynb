# ✅ 2.2 Content Moderation

## 🔎 Meaning

**Content Moderation means checking user input or AI output to block harmful, unsafe, or inappropriate content.**

---

## 🛡️ Used For

```text
Toxic language
Hate speech
Violence
Self-harm content
Sexual content
Abusive messages
```

---

## 🏦 Finance Example

In a finance AI assistant, moderation can block:

```text
Abusive client messages
Harmful instructions
Threatening language
Unsafe or inappropriate content
```

## ✅ One-Line Summary

**Content Moderation blocks harmful or unsafe content before it reaches users or the AI system.**
