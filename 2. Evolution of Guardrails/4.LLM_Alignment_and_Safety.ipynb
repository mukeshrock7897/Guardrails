# ✅ 2.4 LLM Alignment and Safety

## 🧠 Meaning

**LLM Alignment means training and guiding AI models to follow human values, safety rules, and expected behavior.**

---

## 🛡️ Used For

```text id="1cv1ew"
Reducing harmful answers
Improving helpful responses
Following user intent safely
Avoiding biased behavior
Refusing unsafe requests
```

---

## 🏦 Finance Example

In finance, aligned LLMs should avoid:

```text id="cg1cwz"
Guaranteeing returns
Giving risky advice without context
Creating false market claims
Ignoring compliance rules
Using misleading language
```
