# ✅ 3.5 False Positive Risk

## 🚧 Meaning

**False Positive Risk means guardrails may wrongly block a safe or valid user request.**

---

## ❌ Example

A user asks a valid question:

```text
Explain portfolio risk for education purpose.
```

But the guardrail may wrongly treat it as unsafe financial advice.

---

## 🏦 Finance Example

False positives can block:

```text
Valid market research
Internal compliance queries
Portfolio education questions
Approved investment reports
Normal client communication drafts
```

---

## ⭐ Key Points

| Topic              | Importance |
| ------------------ | ---------- |
| Wrong Blocking     | ⭐⭐⭐⭐⭐      |
| User Frustration   | ⭐⭐⭐⭐       |
| Business Delay     | ⭐⭐⭐⭐       |
| Needs Tuning       | ⭐⭐⭐⭐⭐      |
| Human Review Helps | ⭐⭐⭐⭐       |

---

## ✅ One-Line Summary

**False positives happen when guardrails are too strict and block safe, useful, or business-valid requests.**
