# ✅ 3.3 Latency Impact

## 🐢 Meaning

**Latency Impact means guardrails can make AI responses slightly slower because extra checks are added before or after the LLM response.**

---

## ⏱️ Why Latency Increases

```text
Input checking
Policy validation
PII detection
Retrieval validation
Output checking
Human approval
```

---

## 🏦 Finance Example

In finance, response may take longer because AI must check:

```text
Client data access
Investment advice risk
Compliance rules
Approved sources
Risk disclaimer
```

---

## ⭐ Key Points

| Topic                   | Importance |
| ----------------------- | ---------- |
| Extra Safety Checks     | ⭐⭐⭐⭐⭐      |
| Slower Response         | ⭐⭐⭐⭐       |
| Compliance Validation   | ⭐⭐⭐⭐⭐      |
| Human Approval Delay    | ⭐⭐⭐⭐       |
| Production Optimization | ⭐⭐⭐⭐⭐      |

---

## ✅ One-Line Summary

**Guardrails may increase response time, but this delay is acceptable when safety, privacy, and compliance are important.**
