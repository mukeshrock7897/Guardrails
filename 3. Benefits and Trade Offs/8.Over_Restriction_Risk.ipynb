# ✅ 3.8 Over-Restriction Risk

## 📉 Meaning

**Over-Restriction Risk means guardrails become too strict and stop the AI from giving useful answers.**

---

## 🚧 What Can Happen

```text
Valid questions get blocked
Answers become too generic
User experience becomes poor
Business workflows slow down
Human approval increases too much
```

---

## 🏦 Finance Example

AI may refuse too much, even for safe queries like:

```text
Explain diversification in simple words.
Summarize this approved research report.
Compare risk between equity and debt funds.
```

---

## ⭐ Key Points

| Topic                | Importance |
| -------------------- | ---------- |
| Too Many Refusals    | ⭐⭐⭐⭐       |
| Poor User Experience | ⭐⭐⭐⭐       |
| Business Delay       | ⭐⭐⭐⭐       |
| Needs Proper Tuning  | ⭐⭐⭐⭐⭐      |
| Balance is Important | ⭐⭐⭐⭐⭐      |

---

## ✅ One-Line Summary

**Guardrails should protect AI without making it overly restrictive or unusable.**
