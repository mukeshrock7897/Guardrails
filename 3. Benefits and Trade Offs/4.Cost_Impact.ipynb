# ✅ 3.4 Cost Impact

## 💸 Meaning

**Cost Impact means guardrails can increase total AI cost because extra checks, tools, models, and monitoring are used.**

---

## 💰 Why Cost Increases

```text
Input validation
Output validation
PII detection
Hallucination checks
Extra model calls
Monitoring and logging
Human review process
```

---

## 🏦 Finance Example

In finance, cost can increase because systems need:

```text
Compliance checks
Audit logs
Approved source validation
Human approval workflow
Risk monitoring dashboard
```

---

## ⭐ Key Points

| Topic                  | Importance |
| ---------------------- | ---------- |
| Extra Model Calls      | ⭐⭐⭐⭐       |
| Monitoring Cost        | ⭐⭐⭐⭐       |
| Compliance Cost        | ⭐⭐⭐⭐⭐      |
| Human Review Cost      | ⭐⭐⭐⭐       |
| Production Safety Cost | ⭐⭐⭐⭐⭐      |

---

## ✅ One-Line Summary

**Guardrails add cost, but they reduce bigger risks like data leakage, compliance failure, wrong advice, and unsafe AI actions.**
