# ✅ 3.7 Maintenance Effort

## 🔧 Meaning

**Maintenance Effort means guardrails must be regularly updated to stay useful, accurate, and safe.**

---

## 🔄 Why Maintenance is Needed

```text
New risks appear
Policies change
Models change
User behavior changes
Compliance rules change
Business rules change
```

---

## 🏦 Finance Example

Finance guardrails need updates when:

```text
New investment products are added
Compliance policy changes
Approved data sources change
Risk disclosure rules change
New fraud patterns appear
```

---

## ⭐ Key Points

| Topic              | Importance |
| ------------------ | ---------- |
| Policy Updates     | ⭐⭐⭐⭐⭐      |
| Risk Updates       | ⭐⭐⭐⭐⭐      |
| Model Changes      | ⭐⭐⭐⭐       |
| Compliance Updates | ⭐⭐⭐⭐⭐      |
| Continuous Testing | ⭐⭐⭐⭐⭐      |

---

## ✅ One-Line Summary

**Guardrails are not one-time setup; they need continuous updates, testing, and monitoring.**
