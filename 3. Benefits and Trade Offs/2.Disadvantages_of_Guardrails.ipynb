# ✅ 3.2 Disadvantages of Guardrails

## ⚠️ Meaning

**Guardrails improve safety, but they can add extra cost, delay, and complexity to AI systems.**

---

## ❌ Main Disadvantages

```text
Increase response time
Add extra cost
May block valid questions
May miss risky questions
Need regular updates
Make system design more complex
```

---

## 🏦 Finance Example

In finance, guardrails may sometimes:

```text
Block a valid investment research query
Delay answers due to compliance checks
Require human approval for high-risk actions
Increase monitoring and audit effort
```

## ✅ One-Line Summary

**Guardrails are necessary, but they must be balanced carefully so AI remains safe without becoming slow or over-restricted.**
