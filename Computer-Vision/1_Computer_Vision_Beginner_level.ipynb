**What is Computer Vision?**
* Computer Vision is a field of artificial intelligence that enables computers to interpret and make decisions based on visual data from the world. It involves acquiring, processing, analyzing, and understanding images and videos.

**Example:** Recognizing objects in an image, such as detecting cars and pedestrians in a street scene.

**History of Computer Vision**
* **1960s:** Early research focused on image analysis and pattern recognition.
* **1970s:** Development of algorithms for edge detection, segmentation, and feature extraction.
* **1980s:** Introduction of machine learning techniques in vision tasks.
* **1990s:** Progress in real-time image processing and face detection.
* **2000s:** Rise of deep learning and neural networks, leading to significant advancements.

**Example:** The Viola-Jones algorithm for real-time face detection was developed in the early 2000s.

**How Does Computer Vision Work?**
* Computer Vision works by using algorithms and models to interpret visual data. It involves several steps:
* **Image Acquisition:** Capturing images or videos using cameras or sensors.
* **Preprocessing:** Enhancing images by adjusting contrast, removing noise, etc.
* **Feature Extraction:** Identifying important features such as edges, corners, textures, etc.
* **Analysis:** Using algorithms to analyze features and make decisions.

image.png

**Example:** Detecting edges in an image using the Canny edge detection algorithm.

**Computer Vision Techniques**
* **Image Classification:** Assigning a label to an image.
* **Object Detection:** Locating and identifying objects within an image.
* **Image Segmentation:** Partitioning an image into multiple segments.
* **Facial Recognition:** Identifying or verifying a person from an image or video.
* **Motion Analysis:** Analyzing the movement of objects in a video.

**Example:** Detecting cars and pedestrians in an image using a pre-trained YOLO model.

**CV vs. Image Processing**
* Image Processing: Focuses on enhancing images for better visual quality. Techniques include filtering, transformations, etc.

* **Computer Vision:** Involves understanding and interpreting visual data to make decisions.

**Example:** Image processing might involve removing noise from an image, while computer vision could involve recognizing objects in the image.

**Understanding Images and Videos as Data**
* Images and videos are represented as arrays of pixel values. Each pixel in an image has an intensity value, and for color images, it has values for red, green, and blue channels.

### **Example:**



```python
import cv2
import matplotlib.pyplot as plt
import requests  # Import requests to download the image

# Download the image
url = 'https://www.color-hex.com/palettes/18912.png'
response = requests.get(url)
with open('example.jpg', 'wb') as f:
    f.write(response.content)

# Load the image from the local file
image = cv2.imread('example.jpg')

# Check if image loading was successful
if image is None:
    print("Error: Could not load image. Check the file path.")
else:
    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.show()

    # Print image dimensions
    print(f"Image dimensions: {image.shape}")
```

* **Resolution:** The number of pixels in an image, typically represented as width x height.
* **Color Spaces:** Representations of colors, such as RGB, HSV, and Grayscale.

**Example:**

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

**Basic Image Operations**

* Reading and displaying images is the first step in any image processing task.

**Example:**

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Import the patch

# Reading an image
image = cv2.imread('example.jpg')

# Displaying the image using OpenCV
cv2_imshow(image) # Use the patched function
cv2.waitKey(0)
cv2.destroyAllWindows()

# Displaying the image using Matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.show()
```

**Basic Image Transformations: Resizing, Cropping, Rotation**
* **Resizing:** Changing the dimensions of the image.
* **Cropping:** Cutting out a portion of the image.
* **Rotation:** Rotating the image by a certain angle.

**Example:**

```python
# Resizing the image
resized_image = cv2.resize(image, (200, 200))

# Cropping the image
cropped_image = image[50:200, 100:300]

# Rotating the image
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, matrix, (w, h))

# Displaying the transformations
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')

plt.show()
```

* **Image Filtering:** Smoothing, Sharpening, Edge Detection
* **Smoothing (Blurring):** Reducing noise and details.
* **Sharpening:** Enhancing edges in the image.
* **Edge Detection:** Detecting boundaries of objects.

**Example:**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np # Import NumPy

# Smoothing (Blurring) the image
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Sharpening the image
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) # Now you can use np.array
sharpened_image = cv2.filter2D(image, -1, kernel)

# Edge detection using Canny
edges = cv2.Canny(image, 100, 200)

# Displaying the filtered images
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Edges')

plt.show()
```

Handling Mouse Events
Example:

```python
# Callback function for mouse events
import cv2 # Import the cv2 module

def draw_circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(image, (x, y), 20, (255, 0, 0), -1)

# Creating a window and setting a mouse callback
cv2.namedWindow('Mouse Events')
cv2.setMouseCallback('Mouse Events', draw_circle)

while True:
    cv2.imshow('Mouse Events', image)
    if cv2.waitKey(20) & 0xFF == 27:  # Press 'Esc' to exit
        break

cv2.destroyAllWindows()
```

## **Image Histograms**
**Understanding Histograms**
* An image histogram is a graphical representation of the pixel intensity distribution in an image. It plots the number of pixels for each intensity value (0 to 255 for grayscale images).

**Example:**

```python
import cv2
import matplotlib.pyplot as plt

# Load an image in grayscale
image = cv2.imread('example.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate histogram
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot histogram
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

**Histogram Equalization**
* Histogram equalization enhances the contrast of an image by spreading out the most frequent intensity values. This improves the visibility of details in the image.

**Example:**

```python
# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.show()

# Plot histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.plot(cv2.calcHist([image], [0], None, [256], [0, 256]))
plt.xlim([0, 256])

plt.subplot(1, 2, 2)
plt.title('Equalized Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.plot(cv2.calcHist([equalized_image], [0], None, [256], [0, 256]))
plt.xlim([0, 256])

plt.show()
```

**Contrast Limited Adaptive Histogram Equalization (CLAHE)**
* CLAHE is an advanced form of histogram equalization that divides the image into small blocks (tiles) and applies histogram equalization to each block. This technique prevents over-amplification of noise in homogeneous areas.

**Example:**

```python
# Create a CLAHE object
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Apply CLAHE to the grayscale image
clahe_image = clahe.apply(image)

# Display the original and CLAHE images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(clahe_image, cmap='gray')
plt.title('CLAHE Image')

plt.show()

# Plot histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.plot(cv2.calcHist([image], [0], None, [256], [0, 256]))
plt.xlim([0, 256])

plt.subplot(1, 2, 2)
plt.title('CLAHE Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.plot(cv2.calcHist([clahe_image], [0], None, [256], [0, 256]))
plt.xlim([0, 256])

plt.show()
```

**Summary**
* **Histograms:** Represent the frequency distribution of pixel intensities in an image.
* **Histogram Equalization:** Enhances the contrast of an image by spreading out intensity values.
* **CLAHE:** Applies adaptive histogram equalization to small regions of the image, improving local contrast and reducing noise amplification.
