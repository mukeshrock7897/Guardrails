#**1. Introduction to Time Series Analysis**

**What is time series data?**

* Time series data is a sequence of observations collected at regular time intervals. Each observation is associated with a timestamp.

**Types of Time Series Analysis**

*There are two main types:*
* **I-Descriptive analysis:** Understanding patterns and trends in the data.
* **II-Predictive analysis:** Using past data to forecast future values.

**I-Descriptive Analysis**

* **Exploratory Data Analysis (EDA):** This initial phase involves summarizing and visualizing the data to understand its characteristics. It includes techniques like:

* **Descriptive statistics:** Calculating measures like mean, median, standard deviation, and range to get a sense of the central tendency and spread of the data.

* **Visualization:** Creating time series plots, histograms, and other charts to explore trends, seasonality, and potential outliers.

* **Autocorrelation and Partial Autocorrelation:** These measures quantify the relationship between a time series and itself at different lags, helping identify patterns and potential forecasting models.

* **Time Series Decomposition:** Techniques like Seasonal-Trend decomposition by Loess (STL) separate the time series data into its trend, seasonal, and noise components. This helps understand the underlying structure of the data and isolate factors influencing it.

* **Feature Engineering:** This process involves creating new features from the existing data to potentially improve the performance of forecasting models. Examples include lags (past values), moving averages, and ratios.

**II-Predictive Analysis**

* **Forecasting:** This is the core objective of predictive analysis, where models are built to predict future values based on historical data. Common forecasting methods include:

* **Statistical methods:** These models rely on statistical relationships between past and future values. Examples include:
   * **Moving Average models:** Taking the average of a specific number of past data points to predict the next value.
   * **ARIMA (Autoregressive Integrated Moving Average) models:** A powerful statistical method for stationary data forecasting that considers past values (AR), differencing to achieve stationarity (I), and past forecast errors (MA).
* **Machine Learning models:** These models learn complex patterns from the data to make predictions. Examples include:
   * **Recurrent Neural Networks (RNNs):** like Long Short-Term Memory (LSTM) networks: These models are adept at capturing long-term dependencies in time series data.
   * **Convolutional Neural Networks (CNNs):** If your time series data has inherent spatial patterns (e.g., sensor data), CNNs can be effective for forecasting.
   
* **Backtesting and Validation:**  Evaluating the performance of forecasting models is crucial. This involves comparing the predicted values with actual future values using metrics like Mean Squared Error (MSE) or Root Mean Squared Error (RMSE) to assess accuracy. Backtesting involves applying the model to historical data not used for training to gauge its generalizability.

* **Probabilistic Forecasting:**  Instead of a single point prediction, probabilistic forecasting provides a range of possible future values and their associated probabilities. This offers a more comprehensive picture of potential outcomes.

**Examples of time series data**

* Stock prices
* Weather data (temperature, rainfall)
* Website traffic
* Sensor readings (e.g., temperature in a room, heart rate)
* Sales figures
* economic indicators


**Importance of time series analysis**

* Time series analysis helps in understanding past behavior, predicting future trends, and making informed decisions.

**Applications**

* Finance (stock price prediction, risk analysis)
* Supply chain management (demand forecasting)
* Energy forecasting (renewable energy production)
* Healthcare (patient monitoring, disease outbreak prediction)
* Anomaly detection (finding the unusual patterns in your data over time.)
* Resource Planning (the project managers use software to categorize the team according to the skill set and resources they will be using.)
* Weather prediction


**Advantages**

* Provides insights into trends and patterns over time.
* Uncovers hidden patterns in data.
* Enables better planning and resource allocation.
* Automates tasks like forecasting future values.
* Facilitates forecasting and prediction of future outcomes.


**Disadvantages**

* Complexity in handling irregularities and noise.
* Accuracy of forecasts can be limited by randomness in data.
* Requires careful data cleaning and preparation.
* Models can be complex and require expertise to build and interpret.

#**2. Basic Concepts**

**Time series components: trend, seasonality, cyclicity, and irregularity**

* **Trend:** Long-term increase or decrease in data.
* **Seasonality:** Regular pattern that repeats at fixed intervals.
* **Cyclicity:** Trends with no fixed repetition period.
* **Irregularity (or noise):** Random variations in data.


**Stationarity: concept and importance**

* **Stationarity:** A time series is stationary if its statistical properties like mean and variance remain constant over time.

* **Importance:** Stationarity simplifies modeling and forecasting.

**Autocorrelation and Partial Autocorrelation**

* **Autocorrelation:** Correlation of a time series with a lagged version of itself.

* **Partial Autocorrelation:** Direct correlation between two time points, accounting for the indirect effect of other time points.


#**3. Time Series Visualization**

**Plotting time series data**



```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame (replace this with your actual data loading)
data = {'value': [5,10, 20, 15, 25, 30,35,40,45,47,48,50,55,60,65,67,69,73,76,78,79,82,80,84,81,85,88,91,94,96,98,99]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['value'], label='Time Series Data')  # 'df' is now defined
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.legend()
plt.show()
```

**Decomposing time series into components**


```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose time series into trend, seasonal, and residual components
result = seasonal_decompose(df['value'], model='additive', period=12)
result.plot()
plt.show()
```

**Visualizing trends, seasonality, and noise**

```python
# Plot trend component
plt.figure(figsize=(10, 4))
plt.plot(result.trend)
plt.title('Trend Component')
plt.show()

# Plot seasonal component
plt.figure(figsize=(10, 4))
plt.plot(result.seasonal)
plt.title('Seasonal Component')
plt.show()

# Plot residual component
plt.figure(figsize=(10, 4))
plt.plot(result.resid)
plt.title('Residual Component')
plt.show()
```

#**4. Simple Models**

**Mean and Naive methods**

```python
# Mean method
mean_value = df['value'].mean()

# Naive method (last observed value)
naive_value = df['value'].iloc[-1]

print(f"Mean Method: {mean_value}")
print(f"Naive Method: {naive_value}")
```

**Moving averages**

```python
# Simple Moving Average (SMA)
sma = df['value'].rolling(window=3).mean()

# Exponential Moving Average (EMA)
ema = df['value'].ewm(span=3, adjust=False).mean()

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['value'], label='Original Data')
plt.plot(df.index, sma, label='SMA (window=3)')
plt.plot(df.index, ema, label='EMA (span=3)')
plt.xlabel('Date')
plt.legend()
plt.ylabel('Value')
plt.title('Moving Averages')
plt.legend()
plt.show()
```

#**5. Basic Statistical Methods**

**Autoregression (AR)**

```python
from statsmodels.tsa.ar_model import AutoReg

# Fit AutoReg model
model_ar = AutoReg(df['value'], lags=2)
model_ar_fit = model_ar.fit()

# Predict using AutoReg model
predictions_ar = model_ar_fit.predict(start=len(df), end=len(df)+10)

print(f"AR Model Predictions: {predictions_ar}")
```

**Moving Average (MA)**

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit MA model
model_ma = ARIMA(df['value'], order=(0, 0, 1))
model_ma_fit = model_ma.fit()

# Predict using MA model
predictions_ma = model_ma_fit.predict(start=len(df), end=len(df)+10)

print(f"MA Model Predictions: {predictions_ma}")
```

**Autoregressive Moving Average (ARMA)**

```python
# Fit ARMA model
model_arma = ARIMA(df['value'], order=(2, 0, 1))
model_arma_fit = model_arma.fit()

# Predict using ARMA model
predictions_arma = model_arma_fit.predict(start=len(df), end=len(df)+10)

print(f"ARMA Model Predictions: {predictions_arma}")
```
